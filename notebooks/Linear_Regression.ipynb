```python
import numpy as np
```

```python
X = 2* np.random.rand(100,1)
```

```python
y=4+3*X+np.random.randn(100, 1)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(X,y,'o')
```

```python
X_b=np.c_[np.ones((100,1)),X]
```

```python
theta_best=np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
theta_best
```

```python
X_new=np.array([[0],[2]])
```

```python
X_new_b=np.c_[np.ones((2,1)),X_new]
```

```python
X_new_b
```

```python
theta_best
```

```python
y_predict=X_new_b.dot(theta_best)
```

```python
plt.plot(X_new,y_predict,'r-')
plt.plot(X,y,'o')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X,y)
```

```python
yhat_lr=lr.predict(X_new)
```

```python
plt.plot(X_new,yhat_lr,'r-')
plt.plot(X,y,'o')
plt.show()
```
